### Loading Standard Data from sklearn

```python
from sklearn import datasets
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
iris = datasets.load_iris()
# iris.feature_names  # type: ignore , provides labels for dataset
# iris.data #type: ignore , provides data for dataset
df = pd.DataFrame(data=iris.data, columns=iris.feature_names) # type: ignore
df['target'] = iris['target']  # type: ignore
df.head()
```

### Data Reduction using PCA

```python
pca = PCA(2)
df = pca.fit_transform(df)
explained_variance=pca.explained_variance_ratio_ # type: ignore
explained_variance  # type: ignore
plt.bar(range(2), explained_variance, alpha=0.5, align='center')
```

```python
y=iris.target
for i in set(y):
    plt.scatter(df[y == i, 0], df[y == i, 1])

# df[y == 1, 0]
plt.show()
```

## Normalization

Normalization is generally required when we are dealing with attributes on a different scale, otherwise, it may lead to a dilution in effectiveness of an important equally important attribute(on lower scale) because of other attribute having values on larger scale.

Methods of Data Normalization 

    Decimal Scaling:
        v'=v/10^j
        j is smallest value for which v'<1

    Min-Max Normalization
        v'=[(v-min)/(max-min)]*(newMax-newMin)+newMin
        newMax,newMin is the range for new output
        
    z-Score Normalization(zero-mean Normalization)
        v'=(v-mean)/std


## Binning

1. Sort Arrays eg. b=np.sort(b)
2. Create Bins eg. bin=np.zeros((30,5))

Equal Frequency Binning: bins have an equal frequency.

    Input:[5, 10, 11, 13, 15, 35, 50, 55, 72, 92, 204, 215] 

    Output:
    [5, 10, 11, 13]
    [15, 35, 50, 55]
    [72, 92, 204, 215]

Equal Width Binning : bins have equal width with a range of each bin are defined as [min + w], [min + 2w] . [min + nw] where w = (max  min) / (no of bins).

    Input: [5, 10, 11, 13, 15, 35, 50, 55, 72, 92, 204, 215]
        (215-5)/3=210/3=70
        min+70
        min+70+70
        min+70+70+70

    Output:
    [5, 10, 11, 13, 15, 35, 50, 55, 72]
    [92]
    [204, 215]

```python
import numpy as np
import pandas as pd
arr = pd.DataFrame([[180000, 110, 18.9, 1400],
[360000, 905, 23.4, 1800],
[230000, 230, 14.0, 1300],
[60000, 450, 13.5, 1500]]
)

arr.head()
```

### Z-Score Normalisation

```python

#  z-score:
#  v'=(v-mean)/std

for i in arr.columns:
    arr[i]=(arr[i]-arr[i].mean())/arr[i].std()

arr.head()
```

Method 2 : Z score

```python
import scipy.stats as stats
arr['zscore']=stats.zscore(arr[0])
arr
```

```python
# Decimal Scaling
#  v'=v/10**j
toArray=np.array(arr)
numb=toArray.max()

t=0;
while(numb>1):
    numb=numb/10
    t=t+1
print(numb,t)

for i in arr.columns:
    arr[i]=arr[i]/10

arr.head()
```
